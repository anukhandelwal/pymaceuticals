## Screening for potential treatments to squamous cell carcinoma (SCC), a commonly occurring form of skin cancer. Animal study with 250 mice treated through a variety of drug regimes over the course of 45 days, and their physiological responses monitored over the course of that time. 

## Data Analysis to show how four treatments (Capomulin, Infubinol, Ketapril, and Placebo) compare:


### Capomulin was the winner out of the selected 3 drugs for the treatment. Capomulin showed the highest overal
Tumor volume reduction by 19% compared to Infubinol, Ketapril, and placebo, as shown by the "Tumor Change over 45 days" bar chart. 
1.  Tumor Response To Treatment chart :
    Capomulin showed the tumor volumereduction over the course of 45 days while Infubinol, Ketapril, and placebo 
    in tumor volume reduction.
2. Metastatic Spread during treatment chart:
    Capomulin showed the least growth of metstatic sites compared to Infubinol, Ketapril, and placebo
    
3. Mice Survival during Treatment chart:
    Capomulin showed the highest mice survival rate duringtreatment compared to Infubinol, Ketapril, and placebo
    
4. Tumor Changeover 45 days bar chart:
    Capomulin showed the maximum reduction by volume % during treatment compared to Infubinol, Ketapril, and placebo
 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
# Read the raw data files provided

clinical_trial_data = pd.read_csv('raw_data/clinicaltrial_data.csv')
mouse_drug_data = pd.read_csv('raw_data/mouse_drug_data.csv')
```

```python
#CLinical Trial Data data
clinical_trial_data.head()
```

```python
# Mouse drug data
mouse_drug_data.head()
```

#### Merge the Clinical_Trial_Data and Mouse_Drug_data

```python
clinical_data=pd.merge(clinical_trial_data, mouse_drug_data,on ="Mouse ID")
```

```python
clinical_data.head()
```

```python
clinical_data.info()
```

##  how the tumor volume changes over time for each treatment.

```python
# For all rows create a dataframe with columns "Timepoint" " Tumor Volume (mm3)" and "Drug"

tumor_vol_change = clinical_data.loc[:,["Drug", "Timepoint", "Tumor Volume (mm3)"]]
tumor_vol_change.head()
```

```python
# Group the dataset with "Drug" and then "Timepoint", getting mean of timepoint for each drug

tumor_vol_change.groupby("Drug")["Timepoint"].count()
tumor_drug_response = tumor_vol_change.groupby(["Drug", "Timepoint"])
tumor_mean_change = tumor_drug_response["Tumor Volume (mm3)"].mean().to_frame()
tumor_mean_change.head()
```

```python
Tumor_Mean_Change = pd.pivot_table(clinical_data, index='Timepoint', columns='Drug', values='Tumor Volume (mm3)', aggfunc = np.mean)
Tumor_Mean_Change
```

```python
Tumor_Mean_Volume = Tumor_Mean_Change.mean()
Tumor_Mean_Volume
```

```python
Timepoint = Tumor_Mean_Change.index
plt.figure(figsize=(12,8))
plt.errorbar(Timepoint, Tumor_Mean_Change['Capomulin'],yerr=Tumor_Mean_Change['Capomulin'].sem(),color='red', marker ='o', linestyle='--', label="Capomulin")
plt.errorbar(Timepoint, Tumor_Mean_Change['Infubinol'],yerr=Tumor_Mean_Change['Infubinol'].sem(), color='blue', marker ='^', linestyle='--', label="Infubinol")
plt.errorbar(Timepoint, Tumor_Mean_Change['Ketapril'],yerr=Tumor_Mean_Change['Ketapril'].sem(), color='green', marker ='s', linestyle='--', label="Ketapril")
plt.errorbar(Timepoint, Tumor_Mean_Change['Placebo'],yerr=Tumor_Mean_Change['Placebo'].sem(), color='black', marker ='d', linestyle='--', label="Placebo")
plt.gca().set(xlabel = 'Time(Days)', ylabel = 'Tumor Volume (mm3)',title = 'Tumor Response to Treatment',xlim = (0,max(Timepoint)))
plt.legend(loc = 'best', frameon=True)
plt.grid()
plt.show()
```

## Metastatic Response to Treatment

```python
metastatic_response =  clinical_data.groupby(['Drug','Timepoint']).mean()[['Metastatic Sites']]
metastatic_response.head()
```

```python
Metastatic_sites = pd.pivot_table(metastatic_response, index='Timepoint', columns='Drug', values='Metastatic Sites', aggfunc = np.mean)
Metastatic_sites
```

```python
Metastatic = Metastatic_sites.index
plt.figure(figsize=(12,8))

plt.plot(Metastatic, Metastatic_sites['Capomulin'], color='red', marker ='o', linestyle='--', label="Capomulin")
plt.plot(Metastatic, Metastatic_sites['Infubinol'], color = 'blue', marker ='^', linestyle='--', label="Infubinol")
plt.plot(Metastatic, Metastatic_sites['Ketapril'], color='green', marker ='s', linestyle='--', label="Ketapril")
plt.plot(Metastatic, Metastatic_sites['Placebo'], color='black', marker ='d', linestyle='--', label="Placebo")
plt.gca().set(xlabel = 'Treatment Duration (Days)', ylabel = 'Met. Sites',title = 'Metastatic Spread During Treatment',xlim = (0,max(Metastatic)))
plt.legend(loc = 'best', frameon=True)
plt.grid()
plt.show()
```

## the number of mice still alive through the course of treatment (Survival Rate)

```python

mice_survival =  clinical_data.groupby(['Drug','Timepoint']).count()[['Mouse ID']]
mice_survival.head()
```

```python
Survival_pivot = pd.pivot_table(mice_survival, index='Timepoint', columns='Drug', values='Mouse ID', aggfunc = np.mean)
Survival_pivot
```

```python

Survival_percentage = Survival_pivot.copy()
Survival_percentage = round(Survival_percentage.apply(lambda c: c / c.max() * 100, axis=0),2)
Survival_percentage
```

```python
##Survival_rate =  Survival_percentage.index
Survival_rate =  Survival_percentage.index
plt.figure(figsize=(12,8))

plt.plot(Survival_rate, Survival_percentage['Capomulin'], color = 'red', marker ='o', linestyle='--', label="Capomulin")
plt.plot(Survival_rate, Survival_percentage['Infubinol'], color = 'blue', marker ='^', linestyle='--', label="Infubinol")
plt.plot(Survival_rate, Survival_percentage['Ketapril'], color = 'green', marker ='s', linestyle='--', label="Ketapril")
plt.plot(Survival_rate, Survival_percentage['Placebo'], color = 'black', marker ='d', linestyle='--', label="Placebo")
plt.gca().set(xlabel = 'Time (Days)', ylabel = 'Survival Rate(%)',title = 'Survival During Treatment',xlim = (0,max(Survival_rate)))
plt.legend(loc = 'best', frameon=True)
plt.grid()
plt.show()
```

```python
#keep only the drugs we want to examine
clinical_data = clinical_data[(clinical_data["Drug"] == 'Capomulin') |
                      (clinical_data["Drug"] == 'Infubinol') |
                      (clinical_data["Drug"] == 'Ketapril') |
                      (clinical_data["Drug"] == 'Placebo')
                     ]
clinical_data.head()
```

```python
Tumor_Mean_Change = pd.pivot_table(clinical_data, index='Timepoint', columns='Drug', values='Tumor Volume (mm3)', aggfunc = np.mean)
Tumor_Mean_Change.head()
```

```python
##Tumor_volume = Tumor_response.mean()
##Tumor_volume
```

```python
tumor_percentage_x = (((Tumor_Mean_Change.iloc[-1]-Tumor_Mean_Change.iloc[0])/Tumor_Mean_Change.iloc[0])*100).to_frame("% Change")
tumor_percentage_x
```

```python

from math import trunc

##Define plot options, x_label as drug name, y_label as % tumor volume change
x_label=tumor_percentage_x.index
y_label=tumor_percentage_x['% Change']
plt.figure(figsize=(10,6))

## make the bar red if % tuor vol change is positive
## make the bar green if % tumor change volume is negative
## and then graph the bar plot as ax
colors = ['red' if y >=0 else 'green' for y in y_label]
ax = sn.barplot(x_label, y_label, palette=colors)

## for each % tumor change volume, place the % numeric value in the center of the bar
for n, (label, y) in enumerate(zip(x_label, y_label)):
    if y <= 0:
        ax.annotate(
            s='{:d}%'.format(trunc(y)), xy=(n, -3), ha='center',va='center',
            xytext=(0,8), color='w', textcoords='offset points', weight='bold')
    else:
        ax.annotate(
            s='{:d}%'.format(trunc(y)), xy=(n, 0), ha='center',va='center',
            xytext=(0,10), color='w', textcoords='offset points', weight='bold')  

# Bar Plot stye options        
plt.gca().set(xlabel='Drug', ylabel='% Tumor Volume Change', title='Tumor Change Over 45 Day Treatment')
plt.rc('grid', linestyle="--", color='black', linewidth=0.5)
plt.grid(True)
plt.show()
```
